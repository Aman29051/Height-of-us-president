```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\amany\Desktop\president_heights.csv")
data
```

```python
height = data.iloc[:,2]
height.head()
```

```python
print("Mean height : ",height.mean())
print("SD height   : ",height.std())
print("Min height  : ",height.min())
print("Max height  : ",height.max())
```

```python
print("25 percentile  : ", np.percentile(height,25))
print("Median        : ", np.median(height))
print("75 percentile  : ", np.percentile(height,75))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  #set plot style
```

```python
plt.hist(height)
plt.title('Height distribution of US presidents')
plt.xlabel('height(cm)')
plt.ylabel('number')
```
